```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Datos de entrada
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([0, 0, 1, 1, 1])  # Variable de respuesta binaria (0 o 1)

# Ajuste del modelo de regresión logística
model = LogisticRegression()
model.fit(X, y)

# Predicción de nuevas muestras
X_new = np.array([[2.5], [3.5]])
y_pred = model.predict(X_new)
print("Predicciones:", y_pred)

# Graficar los datos y la línea de regresión
plt.scatter(X, y, color='black')
plt.plot(X, model.predict_proba(X)[:, 1], color='blue', linewidth=3)
plt.xlabel('X')
plt.ylabel('Probabilidad de clase 1')
plt.title('Regresión Logística Binomial')
plt.show()
```
