```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos ficticios para ilustrar la relación inversa entre temperatura y cantidad de piratas
temperatura = np.array([20, 23, 24, 25, 26]).reshape(-1, 1)  # Ejemplo de temperatura (en grados Celsius)
cantidad_piratas = np.array([50, 40, 30, 20, 10])  # Ejemplo de cantidad de piratas

# Crear un modelo de regresión lineal
modelo = LinearRegression()

print (temperatura.data.shape, type(temperatura))
print (cantidad_piratas.data.shape, type(cantidad_piratas))

# Entrenar el modelo
modelo.fit(temperatura, cantidad_piratas)

print ("score", modelo.score(temperatura,cantidad_piratas))

# Predecir la cantidad de piratas para las temperaturas dadas
prediccion_piratas = modelo.predict(temperatura)

```

```python
# Graficar el scatter plot y la línea de regresión
plt.figure(figsize=(7, 5))
plt.scatter(cantidad_piratas, temperatura, color='blue', label='Datos')

# El ploteo del predictor
plt.plot(prediccion_piratas, temperatura, color='red', label='Regresión lineal')

# Agregar etiquetas y título
plt.title('Correlación entre temperatura y cantidad de piratas')
plt.xlabel('Cantidad de piratas')
plt.ylabel('Temperatura (°C)')


# Mostrar la leyenda
plt.legend()

# Mostrar el scatter plot con la línea de regresión
plt.show()
```

¿Cual es el score si hago que los puntos coincidan exactamente en algo que se pueda representar con una linea?

```python
temperatura_2 = np.array([20, 25, 30, 35, 40]).reshape(-1, 1)
cantidad_piratas_2 = np.array([40, 35, 30, 25, 20])

modelo_2 = LinearRegression()

modelo_2.fit(temperatura_2, cantidad_piratas_2)

print ("score", modelo_2.score(temperatura_2,cantidad_piratas_2))

prediccion_piratas_2 = modelo_2.predict(temperatura_2)

```

vemos que el score en este caso es 1. Esto indica una precisión perfecta

```python
plt.figure(figsize=(7, 5))
plt.scatter(cantidad_piratas_2, temperatura_2, color='blue', label='Datos')

plt.plot(prediccion_piratas_2, temperatura_2, color='red', label='Regresión lineal')

plt.title('Correlación entre temperatura y cantidad de piratas')
plt.xlabel('Cantidad de piratas')
plt.ylabel('Temperatura (°C)')

plt.legend()

plt.show()
```
