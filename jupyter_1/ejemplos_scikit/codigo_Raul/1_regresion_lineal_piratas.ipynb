```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos ficticios para ilustrar la relación inversa entre temperatura y cantidad de piratas
temperatura = np.array([20, 23, 24, 25, 26]).reshape(-1, 1)  # Ejemplo de temperatura (en grados Celsius)
cantidad_piratas = np.array([50, 40, 30, 20, 10])  # Ejemplo de cantidad de piratas

# Crear un modelo de regresión lineal
modelo = LinearRegression()

# Entrenar el modelo
modelo.fit(temperatura, cantidad_piratas)

print ("El score de este modelo es: ", modelo.score(temperatura,cantidad_piratas))

# Predecir la cantidad de piratas para las temperaturas dadas
prediccion_piratas = modelo.predict(temperatura)

```

## Plot 1: La típica gráfica de iniciación con lo básico

```python
# Graficar el scatter plot y la línea de regresión
plt.figure(figsize=(6, 4))
plt.scatter(cantidad_piratas, temperatura, color='blue', label='Registros pirato-termicos')

# El ploteo del predictor se ve super raro
plt.plot(prediccion_piratas,temperatura, color='red', label='Predictor evolutivo')

# Agregar etiquetas y título
plt.title('Correlación entre temperatura y cantidad de piratas')
plt.xlabel('Cantidad de piratas')
plt.ylabel('Temperatura (°C)')

# Mostrar la leyenda
plt.legend()

# Mostrar el scatter plot con la línea de regresión
plt.show()
```

## Plot 2: Metemos margenes, que se vea un poco más encuadrado

Voy a intentar pintarlo un poco mejor, metiendo margins (no era necesario sacar subplots realmente pero lo dejo asi pq es lo primero que programe)


```python
plt.figure(figsize=(6, 4))

fig, ax = plt.subplots()
ax.plot(prediccion_piratas,temperatura, color='red', label='Predictor evolutivo')
ax.scatter(cantidad_piratas, temperatura, color='blue', label='Registros pirato-termicos')
ax.margins(0.2, 0.2)

# Agregar etiquetas y título
plt.title('Correlación entre temperatura y cantidad de piratas')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Cantidad de piratas')

# Mostrar la leyenda
plt.legend()

# Mostrar el scatter plot con la línea de regresión
plt.show()
```

## Jugar con el Score: ¿Si cocinamos los datos para que coincida la linea de regresión con los mismos el score es 1?

¿Cual es el score si hago que los puntos coincidan exactamente en algo que se pueda representar con una linea?

```python
#Limpia el ploteo anterior
plt.clf()

# Datos ficticios para ilustrar la relación inversa entre temperatura y cantidad de piratas
temperatura_alineada = np.array([20, 22, 24, 26, 28]).reshape(-1, 1)  # Ejemplo de temperatura (en grados Celsius)
cantidad_piratas_alineada = np.array([50, 40, 30, 20, 10])  # Ejemplo de cantidad de piratas

# Crear un modelo de regresión lineal
modelo = LinearRegression()

# Entrenar el modelo
modelo.fit(temperatura_alineada, cantidad_piratas_alineada)

print ("El score de este modelo es: ", modelo.score(temperatura_alineada,cantidad_piratas_alineada))

# Predecir la cantidad de piratas para las temperaturas dadas
prediccion_piratas_alineada = modelo.predict(temperatura_alineada)

# Graficar el scatter plot y la línea de regresión
plt.figure(figsize=(6, 4))
plt.scatter(cantidad_piratas_alineada, temperatura_alineada, color='blue', label='Registros pirato-termicos')

# El ploteo del predictor se ve super raro
plt.plot(prediccion_piratas_alineada,temperatura_alineada, color='red', label='Predictor evolutivo')

# Agregar etiquetas y título
plt.title('Correlación totalmente lineal entre piratas y temperatura')
plt.xlabel('Cantidad de piratas')
plt.ylabel('Temperatura (°C)')

# Mostrar la leyenda
plt.legend()

# Mostrar el scatter plot con la línea de regresión
plt.show()
```

## Plot 3: Que la linea se extienda más alla de los puntos finales y se salga del gráfico

¿Como puedo hacer que la regresion lineal se extienda mas alla de los puntos?

https://stackoverflow.com/questions/22366299/extending-regressions-beyond-data

Basicamente hay que:
1) Generar el plot con datos que se salgan por los lados
2) Eliminar el margin para que no quede un pequeño hueco

```python
#Limpia el ploteo anterior
plt.clf()

# Datos ficticios para ilustrar la relación inversa entre temperatura y cantidad de piratas
temperatura_alineada = np.array([20, 22, 24, 26, 28, 30,32]).reshape(-1, 1)  # Ejemplo de temperatura (en grados Celsius)
cantidad_piratas_alineada = np.array([70, 60, 50, 40, 30, 20, 10])  # Ejemplo de cantidad de piratas

# Crear un modelo de regresión lineal
modelo = LinearRegression()

# Hago un truqui para que la regresion se salga de la grafica
temperatura_alineada_para_scatter = temperatura_alineada[1:-1]
cantidad_piratas_alineada_para_scatter = cantidad_piratas_alineada[1:-1]

# Entrenar el modelo
modelo.fit(temperatura_alineada_para_scatter, cantidad_piratas_alineada_para_scatter)

# Predecir la cantidad de piratas para las temperaturas dadas
prediccion_piratas_alineada_que_se_salga = modelo.predict(temperatura_alineada)

plt.figure(figsize=(6, 4))

fig, ax = plt.subplots()
# Graficar el scatter plot y la línea de regresión
ax.scatter(cantidad_piratas_alineada_para_scatter, temperatura_alineada_para_scatter, color='blue', label='Registros pirato-termicos')

# El ploteo del predictor se ve super raro
ax.plot(prediccion_piratas_alineada_que_se_salga,temperatura_alineada, color='red', label='Predictor evolutivo')
ax.margins(0, 0)
# Agregar etiquetas y título
plt.title('Correlación totalmente lineal entre piratas y temperatura')
plt.xlabel('Cantidad de piratas')
plt.ylabel('Temperatura (°C)')

# Mostrar la leyenda
plt.legend()

# Mostrar el scatter plot con la línea de regresión
plt.show()
```
