```python
import numpy as np
import matplotlib.pyplot as plt

# Coeficientes de la polinomial (por ejemplo, y = ax^2 + bx + c)
coefs = [2, -2, 10]  # Representa la polinomial y = x^2 - 3x + 2

# Definir el rango de valores x
x = np.linspace(-10, 10, 100)  # Aquí especificamos un rango de -10 a 10 con 100 puntos

# Calcular los valores y correspondientes a los valores de x utilizando la polinomial
y = np.polyval(coefs, x)

# Trazar la polinomial
plt.plot(x, y, label='Polinomial')

# Configuraciones adicionales del gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polinomial a partir de coeficientes')
plt.grid(True)
plt.legend()

# Mostrar el gráfico
plt.show()
```
