```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 4, 3.5, 5.5])

# Ajuste de una regresión polinomial de grado 2
degree = 3
coefficients = np.polyfit(x, y, degree)
polynomial = np.poly1d(coefficients)

# Generar puntos para la línea de regresión
x_values = np.linspace(min(x), max(x), 100)
y_values = polynomial(x_values)

# Graficar los datos originales y la línea de regresión
plt.scatter(x, y, label='Datos originales')
plt.plot(x_values, y_values, color='red', label='Regresión polinomial (grado 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión Polinomial')
plt.legend()
plt.grid(True)
plt.show()
```
